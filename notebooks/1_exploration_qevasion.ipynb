# QEvasion Dataset Exploration

This notebook explores the **QEvasion** dataset (Hugging Face: `ailsntua/QEvasion`), used for the project:

*Unmasking Political Question Evasions*

We focus on:
- The structure of the dataset (splits, columns)
- The distributions of:
  - **Clarity labels** (Task 1)
  - **Evasion labels** (Task 2)
- Basic text statistics (length of questions/answers)
- Distributions per president
- The role of `annotator1`, `annotator2`, `annotator3` in the **test** split for Task 2

**Important:**  
For **Task 1 (Clarity)**, the column `clarity_label` is fully annotated in both `train` and `test`.  
For **Task 2 (Evasion)**:
- In `train`, the column `evasion_label` gives a single gold evasion label per example.  
- In `test`, `evasion_label` is empty: instead, we have three annotator columns (`annotator1`, `annotator2`, `annotator3`). Any of these annotator labels is considered correct.

### 5. Text length statistics

We examine the lengths (in words) of:
- `interview_question`
- `interview_answer`
- question + answer combined

This will help later when choosing maximum sequence lengths for transformer models.


```python
def add_lengths(df):
    df = df.copy()
    df["q_len"] = df["interview_question"].fillna("").apply(lambda x: len(x.split()))
    df["a_len"] = df["interview_answer"].fillna("").apply(lambda x: len(x.split()))
    df["qa_len"] = df["q_len"] + df["a_len"]
    return df

train_df = add_lengths(train_df)
test_df  = add_lengths(test_df)

train_df[["q_len", "a_len", "qa_len"]].describe()
```

```python
plt.hist(train_df['q_len'], bins=50, alpha=0.7, edgecolor='black')
plt.title("Question Length (words) – Train")
plt.xlabel("Number of words")
plt.ylabel("Frequency")
plt.axvline(train_df['q_len'].median(), color='red', linestyle='--', label=f"Median: {train_df['q_len'].median():.0f}")
plt.legend()
plt.savefig('../report/figures/question_len_train.png', dpi=300, bbox_inches='tight')
plt.show()

plt.hist(train_df['a_len'], bins=50, alpha=0.7, edgecolor='black', color='orange')
plt.axvline(train_df['a_len'].median(), color='red', linestyle='--', label=f"Median: {train_df['a_len'].median():.0f}")
plt.title("Answer Length (words) – Train")
plt.xlabel("Number of words")
plt.ylabel("Frequency")
plt.legend()
plt.savefig('../report/figures/answer_len_train.png', dpi=300, bbox_inches='tight')
plt.show()

plt.hist(train_df['qa_len'], bins=50, alpha=0.7, edgecolor='black', color='green')
plt.title("Question + Answer Length (words) – Train")
plt.xlabel("Number of words")
plt.ylabel("Frequency")
plt.axvline(train_df['qa_len'].median(), color='red', linestyle='--', label=f"Median: {train_df['qa_len'].median():.0f}")
plt.legend()
plt.savefig('../report/figures/qna_len_train.png', dpi=300, bbox_inches='tight')
plt.show()

length_data = [train_df['q_len'], train_df['a_len'], train_df['qa_len']]
plt.boxplot(length_data, tick_labels=['Question', 'Answer', 'Q+A'])
plt.ylabel('Length (words)')
plt.title('Length Comparison (Box Plot)')
plt.grid(alpha=0.3)
plt.savefig('../report/figures/qna_boxplot_train.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# Print percentiles for model selection
print("\nLength Percentiles (for choosing max sequence length):")
print("\nQuestion + Answer combined:")
for p in [50, 75, 90, 95, 99]:
    val = np.percentile(train_df['qa_len'], p)
    print(f"  {p}th percentile: {val:.0f} words")
```

### 6. Distributions per president

The dataset includes a `president` column (e.g.\ Bush, Obama, Trump, Biden).  
We explore:
- How many examples per president
- How clarity labels are distributed per president
- How evasion labels are distributed per president (on the train split)


```python
print("=== Examples per president - TRAIN ===")
print(train_df["president"].value_counts(), "\n")

print("=== Examples per president - TEST ===")
print(test_df["president"].value_counts())
```

```python
clarity_pres = pd.crosstab(train_df["president"], train_df["clarity_label"])
clarity_pres
```

```python
clarity_pres.plot(kind="bar", stacked=True)
plt.title("Clarity Labels by President (Train)")
plt.ylabel("Count")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.legend(title='Clarity', bbox_to_anchor=(1.05, 1))
plt.savefig('../report/figures/clarity_president_train.png', dpi=300, bbox_inches='tight')
plt.show()

# Evasion by president (top 5 categories)
top_evasion = train_df['evasion_label'].value_counts().head(5).index
evasion_pres = pd.crosstab(
    train_df['president'],
    train_df['evasion_label']
)[top_evasion]
evasion_pres.plot(kind='bar', stacked=True)
plt.title('Top 5 Evasion Labels by President')
plt.xlabel('President')
plt.ylabel('Count')
plt.legend(title='Evasion', bbox_to_anchor=(1.05, 1))
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('../report/figures/top_evasion_president_train.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# President vs Evasion cross-tabulation
evasion_pres = pd.crosstab(train_df["president"], train_df["evasion_label"])
evasion_pres
```

```python
# Clarity vs Evasion cross-tabulation
crosstab = pd.crosstab(train_df['clarity_label'], train_df['evasion_label'], normalize='index').mul(100).round(1)

print("\nClarity vs Evasion Label Correlation (% within each clarity category):")
display(crosstab)

# Visualize as heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(crosstab, annot=True, fmt='.1f', cmap='YlOrRd', cbar_kws={'label': 'Percentage'})
plt.title('Clarity vs Evasion Label Distribution (%)')
plt.xlabel('Evasion Label')
plt.ylabel('Clarity Label')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('../report/figures/clarity_evasion_correlation.png', dpi=300, bbox_inches='tight')
plt.show()
```
